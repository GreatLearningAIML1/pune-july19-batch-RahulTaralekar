   2. sepal width in cm

### Find out the mean and variance for each column but for class column 

```python
for (columnName, columnData) in iris.iteritems():
    if iris[columnName].dtype!="object":
        print ("Mean of column "+ columnName + " is: " + str(iris[columnName].mean()))
        print ("Variance of column " + columnName +" is: " + str(iris[columnName].var()))
```

### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
for (columnName, columnData) in iris.iteritems():
    if iris[columnName].dtype!="object":
        print ("Range of  column "+ columnName + " :" + str(iris[columnName].max()-iris[columnName].min()))
```
