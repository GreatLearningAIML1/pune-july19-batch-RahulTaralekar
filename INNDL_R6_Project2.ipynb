**Lets create a CNN and see how is the model accuracy**

```
# Define Model
    model_cnn = tf.keras.models.Sequential()
    
    #Reshape data from 2D (32,32) to 3D (32, 32, 1)
    model_cnn.add(tf.keras.layers.Reshape((32,32,1),input_shape=(32,32,)))

    model_cnn.add(tf.keras.layers.BatchNormalization())

    # 1st Conv Layer
    model_cnn.add(tf.keras.layers.Conv2D(32,kernel_size=(3,3),activation='relu'))
      
    # Dropout
    model_cnn.add(tf.keras.layers.Dropout(0.2))


    # 2nd Conv Layer
    
    model_cnn.add(tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu'))
    model_cnn.add(tf.keras.layers.BatchNormalization())
    
    # Max Pooling
    model_cnn.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
    # Dropout
    model_cnn.add(tf.keras.layers.Dropout(0.2))

    
     # 3rd Conv Layer
    model_cnn.add(tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu'))
    model_cnn.add(tf.keras.layers.BatchNormalization())
    
    # Max Pooling
    model_cnn.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
    # Dropout
    model_cnn.add(tf.keras.layers.Dropout(0.2))
    
     
     # 4th Conv Layer
    model_cnn.add(tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu'))
    model_cnn.add(tf.keras.layers.BatchNormalization())
    
    # Max Pooling
    model_cnn.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
    # Dropout
    model_cnn.add(tf.keras.layers.Dropout(0.25))
    
    # Fully Connected Layer
    model_cnn.add(tf.keras.layers.Flatten())
    model_cnn.add(tf.keras.layers.Dense(128, activation='relu'))
    model_cnn.add(tf.keras.layers.Dense(256, activation='relu'))
        
    # Prediction Layer
    model_cnn.add(tf.keras.layers.Dense(10, activation = 'softmax'))
    
    # Loss and Optimizer
    model_cnn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
  
  
```

```
#Train the model
model_cnn.fit(x_train_norm,trainY,          
          validation_data=(x_test_norm, testY),
          epochs=10,
          batch_size=32)        
```

```
# model evaluation
score = model_cnn.evaluate(x_test_norm, testY, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```
# predict the model
y_pred = model_cnn.predict_classes(x_test)
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
